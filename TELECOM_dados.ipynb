```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calcular a porcentagem de cada gênero
gender_counts = df['Gênero'].value_counts(normalize=True) * 100
plt.figure(figsize=(10, 5))
sns.barplot(x=gender_counts.index, y=gender_counts.values, palette='pastel')
plt.title('Distribuição de Gênero dos Clientes em Porcentagem')
plt.xlabel('Gênero')
plt.ylabel('Porcentagem (%)')
plt.ylim(0, 100)  # Limitar o eixo y de 0 a 100
for index, value in enumerate(gender_counts.values):
    plt.text(index, value + 1, f'{value:.1f}%', ha='center')
plt.show()

# Calcular a porcentagem de idosos
elderly_counts = df['Idoso'].value_counts(normalize=True) * 100
plt.figure(figsize=(10, 5))
sns.barplot(x=elderly_counts.index, y=elderly_counts.values, palette='pastel')
plt.title('Proporção de Idosos entre os Clientes em Porcentagem')
plt.xlabel('Idoso (1 = Sim, 0 = Não)')
plt.ylabel('Porcentagem (%)')
plt.xticks([0, 1], ['Não', 'Sim'])  # Ajustar os rótulos do eixo x
plt.ylim(0, 100)  # Limitar o eixo y de 0 a 100
for index, value in enumerate(elderly_counts.values):
    plt.text(index, value + 1, f'{value:.1f}%', ha='center')
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df['Faturamento'] = pd.to_numeric(df['Faturamento'], errors='coerce')

df = df.dropna(subset=['Faturamento'])

# Ou pode preencher os NaN com a média
# df['Faturamento'].fillna(df['Faturamento'].mean(), inplace=True)

plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x='Dependentes', y='Faturamento', alpha=0.6)
plt.title('Relação entre Número de Dependentes e Receita Gerada por Cliente')
plt.xlabel('Número de Dependentes')
plt.ylabel('Faturamento')
plt.axhline(y=df['Faturamento'].mean(), color='r', linestyle='--', label='Faturamento Médio')  # Linha média
plt.legend()
plt.grid()
plt.show()
```
